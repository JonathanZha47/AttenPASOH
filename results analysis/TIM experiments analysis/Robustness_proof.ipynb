# Robustness Proof

With gaussian noise (0,1) versus Without gaussian noise(0,1)

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Define the base paths to the experiments
base_path_1 = "/Users/jonathanzha/Desktop/AttenPASOH/results of reviewer/XJTU(Baseline)+noisy results/0-0"
base_path_2 = "/Users/jonathanzha/Desktop/AttenPASOH/results of reviewer/XJTU(Baseline) results/0-0"

experiment = "Experiment1"

# Define paths for both experiments
pred_label_path_1 = os.path.join(base_path_1, experiment, 'pred_label.npy')
true_label_path_1 = os.path.join(base_path_1, experiment, 'true_label.npy')

pred_label_path_2 = os.path.join(base_path_2, experiment, 'pred_label.npy')
true_label_path_2 = os.path.join(base_path_2, experiment, 'true_label.npy')

# Load the labels for both experiments
pred_label_1 = np.load(pred_label_path_1)
true_label_1 = np.load(true_label_path_1)

pred_label_2 = np.load(pred_label_path_2)
true_label_2 = np.load(true_label_path_2)

# Create a left-right layout for scatter plots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Scatter plot for first experiment
axs[0].scatter(true_label_1, pred_label_1, color='green', s=10, alpha=0.6)
axs[0].plot([0.8, 1], [0.8, 1], color='red', linestyle='--', linewidth=2)
axs[0].set_title('True vs Predicted (Noisy Results)', fontsize=16, fontname='Times New Roman')
axs[0].set_xlabel('True Label', fontsize=14)
axs[0].set_ylabel('Predicted Label', fontsize=14)
axs[0].grid(True)

# Scatter plot for second experiment
axs[1].scatter(true_label_2, pred_label_2, color='blue', s=10, alpha=0.6)
axs[1].plot([0.8, 1], [0.8, 1], color='red', linestyle='--', linewidth=2)
axs[1].set_title('True vs Predicted (Baseline Results)', fontsize=16, fontname='Times New Roman')
axs[1].set_xlabel('True Label', fontsize=14)
axs[1].set_ylabel('Predicted Label', fontsize=14)
axs[1].grid(True)

# Adjust the layout and show the plots
plt.tight_layout()
plt.savefig("/Users/jonathanzha/Desktop/AttenPASOH/useful plots/noisyVsNonoisy.png")
plt.show()
```
