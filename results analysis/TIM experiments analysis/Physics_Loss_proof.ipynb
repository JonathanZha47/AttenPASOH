```python
import os
import numpy as np
import pandas as pd
import sys
sys.path.append("/Users/jonathanzha/Desktop/AttenPASOH")
from utils.util import eval_metrix  # Assuming this function exists and is correctly implemented
```

Below is the analysis of baseline wiz the physics loss: ReLU(u2-u1).sum()

```python
# Define the path to the new experiment
experiments_path_wizPhysicsLossY = "/Users/jonathanzha/Desktop/AttenPASOH/results of reviewer/XJTU(Baseline wiz physics loss(y)) results/0-0"
experiments = [f"Experiment{i+1}" for i in range(10)]
```

```python

# Initialize lists to store metrics for each experiment
mse_list, mae_list, mape_list, rmse_list, r2_list, l1_list, l2_list = [], [], [], [], [], [], []

def calculate_metrics(true_label_path, pred_label_path):
    # Load true and predicted labels
    true_label = np.load(true_label_path)
    pred_label = np.load(pred_label_path)
    
    # Calculate metrics using the eval_matrix function
    mse, mae, mape, rmse, r2, l1, l2 = eval_metrix(true_label, pred_label)
    
    return mse, mae, mape, rmse, r2, l1, l2

# Loop over each experiment
for experiment in experiments:
    true_label_path = os.path.join(experiments_path_wizPhysicsLossY, experiment, 'true_label.npy')
    pred_label_path = os.path.join(experiments_path_wizPhysicsLossY, experiment, 'pred_label.npy')
    
    # Calculate metrics for the current experiment
    metrics = calculate_metrics(true_label_path, pred_label_path)
    
    mse, mae, mape, rmse, r2, l1, l2 = metrics
    mse_list.append(mse)
    mae_list.append(mae)
    mape_list.append(mape)
    rmse_list.append(rmse)
    r2_list.append(r2)
    l1_list.append(l1)
    l2_list.append(l2)

# Create a DataFrame to store the metrics
metrics_df = pd.DataFrame({
    'Experiment': experiments,
    'MSE': mse_list,
    'MAE': mae_list,
    'MAPE': mape_list,
    'RMSE': rmse_list,
    'R^2': r2_list,
    'L1 Error': l1_list,
    'L2 Error': l2_list
})

# Calculate the mean values across all numeric columns (excluding 'Experiment')
mean_values = metrics_df.drop(columns=['Experiment']).mean(axis=0)

# Append the mean values to the DataFrame
mean_values_df = pd.DataFrame(mean_values).transpose()
mean_values_df['Experiment'] = 'Mean'
metrics_df = pd.concat([metrics_df, mean_values_df], ignore_index=True)

# Print the results
print(metrics_df)

# Save the results to a CSV file
output_csv_path = os.path.join(experiments_path_wizPhysicsLossY, 'metrics_summary.csv')
metrics_df.to_csv(output_csv_path, index=False)
print(f"Metrics summary saved to {output_csv_path}")
```

```python
import os
import re
import numpy as np
import matplotlib.pyplot as plt

# Define the directories containing the experiments
experiments_path_wizPhysicsLossY = '/Users/jonathanzha/Desktop/AttenPASOH/results of reviewer/XJTU(Baseline wiz physics loss(y)) results/0-0'
experiments_path_wizNoPhyLoss = '/Users/jonathanzha/Desktop/AttenPASOH/results of reviewer/XJTU(Baseline wiz no physics loss) results/0-0'
experiment_dirs = [f'Experiment{i}' for i in range(1, 11)]

# Initialize lists to hold training losses for both paths
train_losses_physics = []
train_losses_no_physics = []

# Regular expressions to match the loss lines
train_loss_pattern = re.compile(r'\[Train\] epoch:(\d+), lr:.* total loss:(\d+\.\d+)')

# Function to extract losses from log files
def extract_losses(base_dir, train_losses):
    for exp in experiment_dirs:
        log_file_path = os.path.join(base_dir, exp, 'logging.txt')
        
        if os.path.exists(log_file_path):
            with open(log_file_path, 'r') as file:
                for line in file:
                    train_match = train_loss_pattern.search(line)
                    
                    if train_match:
                        epoch = int(train_match.group(1))
                        loss = float(train_match.group(2))
                        train_losses.append((epoch, loss))

# Extract training losses for both experiment paths
extract_losses(experiments_path_wizPhysicsLossY, train_losses_physics)
extract_losses(experiments_path_wizNoPhyLoss, train_losses_no_physics)

# Function to aggregate losses by epoch
def aggregate_losses(losses, max_epoch):
    loss_dict = {i: [] for i in range(1, max_epoch + 1)}
    for epoch, loss in losses:
        if epoch <= max_epoch:
            loss_dict[epoch].append(loss)

    return loss_dict

# Aggregate the losses
max_epoch = 50  # Set maximum epoch for plotting
train_loss_dict_physics = aggregate_losses(train_losses_physics, max_epoch)
train_loss_dict_no_physics = aggregate_losses(train_losses_no_physics, max_epoch)

# Calculate min, mean, and max for each epoch for physics loss
train_min_physics = [min(train_loss_dict_physics[i]) if train_loss_dict_physics[i] else np.nan for i in range(1, max_epoch + 1)]
train_mean_physics = [np.mean(train_loss_dict_physics[i]) if train_loss_dict_physics[i] else np.nan for i in range(1, max_epoch + 1)]
train_max_physics = [max(train_loss_dict_physics[i]) if train_loss_dict_physics[i] else np.nan for i in range(1, max_epoch + 1)]

# Calculate min, mean, and max for each epoch for no physics loss
train_min_no_physics = [min(train_loss_dict_no_physics[i]) if train_loss_dict_no_physics[i] else np.nan for i in range(1, max_epoch + 1)]
train_mean_no_physics = [np.mean(train_loss_dict_no_physics[i]) if train_loss_dict_no_physics[i] else np.nan for i in range(1, max_epoch + 1)]
train_max_no_physics = [max(train_loss_dict_no_physics[i]) if train_loss_dict_no_physics[i] else np.nan for i in range(1, max_epoch + 1)]

# Plotting
epochs = np.arange(1, max_epoch + 1)

plt.figure(figsize=(12, 6))

# Training Loss with Physics
plt.fill_between(epochs, train_min_physics, train_max_physics, color='orange', alpha=0.3, label='Training Loss (Physics) Min-Max')
plt.plot(epochs, train_mean_physics, label='Training Loss (Physics) Mean', color='orange', linewidth=2)

# Training Loss without Physics
plt.fill_between(epochs, train_min_no_physics, train_max_no_physics, color='red', alpha=0.3, label='Training Loss (No Physics) Min-Max')
plt.plot(epochs, train_mean_no_physics, label='Training Loss (No Physics) Mean', color='red', linewidth=2)

# Labels and title
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title('Training Losses for Physics vs No Physics')
plt.legend()
plt.grid()

# Save the plot
plt.tight_layout()
plt.savefig(os.path.join(experiments_path_wizNoPhyLoss, 'combined_training_loss_plot.png'))
plt.show()
```
