```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the attention weights
atten_weight_path = "/Users/jonathanzha/Desktop/Meta-Learning-PINN-for-SOH/results/Comparison_Experiment/XJTU-attenPASOH(2C+32+0.25equalweight) results/0-0/Experiment1/atten_weight.npy"
atten_weight = np.load(atten_weight_path)

# Check the shape of the attention weights
print("Shape of attention weights:", atten_weight.shape)  # Should output (4, 8, 3, 3, 3)

# Number of layers, heads, and sequence length
n_layers = atten_weight.shape[0]  # 4 layers
n_heads = atten_weight.shape[1]   # 8 attention heads
seq_length = atten_weight.shape[2]  # 3 (time steps)

# Plot the attention weights for each layer and head
for layer in range(n_layers):
    plt.figure(figsize=(16, 12))  # Adjusting the figure size
    for head in range(n_heads):
        # Attention weights for the current layer and head (shape [seq_length, 3, 3])
        attn_weights = atten_weight[layer][head]

        # Create subplot for each head
        plt.subplot(2, 4, head + 1)  # 2 rows and 4 columns for 8 heads
        for t in range(seq_length):
            # Plot heatmap for each timestep in the sequence
            sns.heatmap(attn_weights[t], cmap='viridis', annot=True, cbar=True)
            plt.title(f'Layer {layer + 1}, Head {head + 1}, Time Step {t + 1}')
    
    plt.suptitle(f'Attention Weights - Layer {layer + 1}')
    plt.tight_layout()
    plt.show()
```
