```python
def denormalize(pred_label, nominal_capacity, min_val, max_val):
    # Step 1: Reverse the Min-Max scaling
    original_value_after_dividing = (pred_label + 1) / 2 * (max_val - min_val) + min_val
    
    # Step 2: Reverse the division by nominal_capacity
    original_capacity = original_value_after_dividing * nominal_capacity
    
    return original_capacity
```
