# Batch size experiment plot

We use 8, 16, 32, 64, 128, 256, 512 for our model to run the experiment. Below is the visualization:

```python
import matplotlib.pyplot as plt

# Data from the mean value of 10 experiments that is done under the 2C batch of the XJTU dataset.
batch_sizes = [8, 16, 32, 64, 128, 256, 512]
mse_values = [0.000058, 0.000045, 0.000038, 0.00005, 0.000112, 0.00017, 0.000352]
mae_values = [0.005768, 0.004878, 0.004498, 0.005173, 0.008073, 0.010177, 0.014901]
mape_values = [0.006352, 0.005403, 0.004983, 0.005716, 0.008851, 0.011109, 0.016219]
rmse_values = [0.007502, 0.006602, 0.006063, 0.006964, 0.010252, 0.012604, 0.018267]
r_square_values = [0.976564, 0.981937, 0.984653, 0.979857, 0.954473, 0.930654, 0.856751]

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 8))

# Left Y-axis for R-squared
ax1.set_xlabel('Batch Size')
ax1.set_ylabel('R-squared')
ax1.plot(batch_sizes, r_square_values, marker='o', color='blue', label='R-squared')
ax1.tick_params(axis='y')
# Set explicit labels for batch sizes
ax1.set_xscale('log')
ax1.set_xticks(batch_sizes)
ax1.get_xaxis().set_major_formatter(plt.ScalarFormatter())  # Force explicit values instead of log scale formatting
ax1.grid(True)

# Right Y-axis for MAE, MAPE, RMSE
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Metric Value', color='black')
ax2.plot(batch_sizes, mae_values, marker='o', linestyle='-', color='green', label='MAE')
ax2.plot(batch_sizes, mape_values, marker='o', linestyle='-', color='orange', label='MAPE')
ax2.plot(batch_sizes, rmse_values, marker='o', linestyle='-', color='purple', label='RMSE')
ax2.tick_params(axis='y', labelcolor='black')

# Adding legends for both y-axes
fig.tight_layout()  # To prevent overlap
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('R-squared vs Batch Size (Left Y-axis) and Other Metrics (Right Y-axis)')
plt.show()
```
