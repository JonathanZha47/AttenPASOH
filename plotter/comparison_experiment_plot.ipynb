```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Define the base path to the experiments
base_path = "/Users/jonathanzha/Desktop/Meta-Learning-PINN-for-SOH/results/Comparison_Experiment/XJTU-attenPASOH(2C+32+0.25equalweight) results/0-0"

# Define the experiments (you can modify the range if there are more/less than 10 experiments)
experiments = [f"Experiment{i+1}" for i in range(10)]

# Initialize lists to store all true and predicted labels
all_true_labels = []
all_pred_labels = []

# Loop through each experiment and load the labels
for experiment in experiments:
    pred_label_path = os.path.join(base_path, experiment, 'pred_label.npy')
    true_label_path = os.path.join(base_path, experiment, 'true_label.npy')
    
    if os.path.exists(pred_label_path) and os.path.exists(true_label_path):
        # Load the labels
        pred_label = np.load(pred_label_path)
        true_label = np.load(true_label_path)
        
        # Append the labels to the respective lists
        all_true_labels.append(true_label)
        all_pred_labels.append(pred_label)
    else:
        print(f"Missing files for {experiment}")

# Concatenate all the labels together
all_true_labels = np.concatenate(all_true_labels)
all_pred_labels = np.concatenate(all_pred_labels)

# Line Plot: True Label vs Predicted Label
plt.figure(figsize=(10, 6))
plt.plot(all_true_labels, label='True Label', color='blue', linestyle='-', linewidth=2)
plt.plot(all_pred_labels, label='Predicted Label', color='red', linestyle='--', linewidth=2)

# Adding title and labels for line plot
plt.title('Combined True Label vs Predicted Label (Atten-PASOH)', fontsize=16)
plt.xlabel('Sample Index', fontsize=14)
plt.ylabel('Label Value', fontsize=14)
plt.legend(loc='best')
plt.grid(True)

# Show the line plot
plt.show()

# Residual Plot: True Label - Predicted Label
plt.figure(figsize=(10, 6))
plt.plot(all_true_labels - all_pred_labels, color='purple', linestyle='-', linewidth=2)
plt.title('residual vs sample points(Atten-PASOH)', fontsize=16)
plt.xlabel('Sample Index', fontsize=14)
plt.ylabel('Residual Value', fontsize=14)
plt.legend(loc='best')
plt.grid(True)
plt.show()


# Scatter Plot: True Label vs Predicted Label
plt.figure(figsize=(10, 6))
plt.scatter(all_true_labels, all_pred_labels, color='green', s=10, alpha=0.6)
plt.plot([0.8, 1], [0.8, 1], color='red', linestyle='--', linewidth=2)

# Adding title and labels for scatter plot
plt.title('True Label vs Predicted Label (Atten-PASOH)', fontsize=16)
plt.xlabel('True Label', fontsize=14)
plt.ylabel('Predicted Label', fontsize=14)
plt.grid(True)

# Show the scatter plot
plt.show()
```
